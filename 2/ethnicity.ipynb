```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#read in the csv
race_ethnicity = pd.read_excel("data/building_ethnic_1718.xlsx")
```

```python
#visualize 
race_ethnicity.head()
```

```python
# filter for the three schools being studied
bldg_irns = [729, 24687, 32060]
three_schools = race_ethnicity[race_ethnicity['Building IRN'].isin(bldg_irns)]
```

```python
three_schools
```

```python
#replace all NC with NaNs so I can manipulate

three_schools.replace('NC', np.nan, inplace=True)
```

```python
#make sure it worked

three_schools
```

```python
#drop all columns with all NA values

three_schools.dropna(axis=1, how='all', inplace=True)
```

```python
three_schools
```

```python
#list of remaining columns
three_schools.columns.values
```

```python
# it seems like there are a lot of race and ethnicity rows for the schools that are empty or are not measured. want to get rid of rows that don't have info
#so its not so cluttered
# use subset, thresh and axis parameters of dropna to accomplish this 
#basically, if 16 of the columns listed here are null in a row, drop the row

three_schools.dropna(axis=0, subset=['ELA 3rd Grade 2017-2018 % Proficient or above', \
                                      'Math 3rd Grade 2017-2018 % Proficient or above', \
                                      'ELA 4th Grade 2017-2018 % Proficient or above', \
                                      'Math 4th Grade 2017-2018 % Proficient or above', \
                                      'ELA 5th Grade 2017-2018 % Proficient or above', \
                                      'Math 5th Grade 2017-2018 % Proficient or above', \
                                      'Science 5th Grade 2017-2018 % Proficient or above', \
                                      'ELA 6th Grade 2017-2018 % Proficient or above', \
                                      'Math 6th Grade 2017-2018 % Proficient or above', \
                                      'ELA 7th Grade 2017-2018 % Proficient or above', \
                                      'Math 7th Grade 2017-2018 % Proficient or above', \
                                      'ELA 8th Grade 2017-2018 % Proficient or above', \
                                      'Math 8th Grade 2017-2018 % Proficient or above', \
                                      'Science 8th Grade 2017-2018 % Proficient or above', \
                                      'Attendance Rate 2017-2018', \
                                      'Enrollment 2017-2018', \
                                      'Percent of Total Enrollment 2017-2018'], thresh=16) 
```

```python
three_schools.to_csv("data/reduced_ethnicity_dataset.csv")
#not much to compare here in terms of scores across ethnic or racial lines, but i do think the fluctuations in % proficient or above are interesting
# of course, this is measuring different cohorts and is just a snapshot, this does not show how the same cohort's proficiency changes year over year
```
